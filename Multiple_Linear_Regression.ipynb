```
%matplotlib inline
from fastai.basics import *
```

```
n=100
x = torch.ones(n,3) # type: torch.Tensor
x[:,:2].uniform_(-1.,1) # apply uniform_ on 1st column
x[:5]
```

```
a = tensor(3.,-4,2)
a
```

```
# See what our dataset looks like
y = x@a # matrix multiplication
p1 = x[:,0]*3.
p2 = x[:,1]*-4.
p3 = x[:,2]*2.

fig, (ax1,ax2,ax3) = plt.subplots(1,3, figsize=(16,4))
ax1.scatter(x[:,0], y)
ax2.scatter(x[:,1], y)
ax3.scatter(p1+p2, y)
```

Suppose we don't know that the target parameters is `a = tensor(3.,-4,2)`

Let's optimize the initial parameters

```
# Set initial parameters
a = nn.Parameter(tensor(30.,10,4))
a
```

```
# See what our dataset looks like
y_hat = x@a # matrix multiplication
y_hat = y_hat.detach()
fig, (ax1,ax2) = plt.subplots(1,2, figsize=(16,4))
ax1.scatter(x[:,0], y, label="actual")
ax1.scatter(x[:,0], y_hat, label="preds")
ax1.legend(loc="upper left")
ax2.scatter(x[:,1], y, label="actual")
ax2.scatter(x[:,1], y_hat, label="preds")
ax2.legend(loc="upper left")
```

```
# loss function
def mse(y_hat, y): return ((y_hat-y)**2).mean()
```

```
def update(n, lr):
    '''updates vector a'''
    losses = [] # tracks losses history
    for i in range(n):
        y_hat = x@a
        loss = mse(y_hat, y) # y_hat -> predictions, y -> actual value
        losses.append(loss)

        if i % 10 == 0: 
            print(loss) # just prints stuff, DOESN'T AFFECT THE ALGO
            
        loss.backward() # computes gradient of current tensor wrt to graph leaves (need to dig into this)

        with torch.no_grad():
            a.sub_(lr * a.grad) # subtracts learning rate * gradient to the parameters (inplace)
            a.grad.zero_() # reset grad to zero because torch accumulates grad.. you don't want to reuse it
    
    #plt.plot(list(range(0,len(losses))), losses)
```

```
# Update and then see what our dataset looks like

update(10, 1e-1)

y_hat = x@a # matrix multiplication

fig, (ax1,ax2,ax3) = plt.subplots(1,3, figsize=(16,4))
ax1.scatter(x[:,0], y, label="actual")
ax1.scatter(x[:,0], y_hat.detach(), label="preds")
ax1.legend(loc="upper left")
ax2.scatter(x[:,1], y, label="actual")
ax2.scatter(x[:,1], y_hat.detach(), label="preds")
ax2.legend(loc="upper left")
ax3.scatter(y_hat.detach(),y)
```
